```python
import pathlib
import pandas as pd

from spacepy import pycdf
```

```python
cdfs = [pycdf.CDF(str(path)) for path in pathlib.Path('../data').rglob('*.cdf') if str(path.name)[8:10] == 'hk']
```

```python
cdfs[0]
```

```python
def cdf_fields(cdf: pycdf.CDF):
    return list(map(lambda f: f[0], cdf.items()))
```

```python
def to_dataframe(cdf: pycdf.CDF):
    return pd.DataFrame({
        field: cdf[field] for field in cdf_fields(cdf)
    })
```

```python
df_100 = to_dataframe(cdfs[100])
df_100
```

```python
df = pd.concat(to_dataframe(cdf) for cdf in cdfs)
```
