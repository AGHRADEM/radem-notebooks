# SREM

```python
import requests as r
import datetime as datetime
```

```python
START = (2002, 10, 17)
END = datetime.datetime.now().date().year, datetime.datetime.now().date().month, datetime.datetime.now().date().day
```

```python
END, START
```

```python
def get(year, month, day):
    uri = f'http://srem.psi.ch/datarepo/V0/irem/{year:04}/IREM_PACC_{year:04}{month:02}{day:02}.cdf.gz'
    res = r.get(uri)
    return uri, res
```

```python
def listall():
    for year in range(START[0], END[0]+1):
        for month in range(1, 13):
            for day in range(1, 32):
                uri, res = get(year, month, day)
                print(uri, res.status_code, flush=True)
                

a, b = listall()
print(a, b.status_code)
```
